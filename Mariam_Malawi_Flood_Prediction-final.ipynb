## Data Cleaning

#### Separating the train and test sets 

```python
# Creating lists of columns to be used in separating the dataframe into training and testing datasets
# Using a for loop for efficiency
#
precip_features_2019 = []
precip_features_2015 = []
for col in df.columns:
  if '2019' in col:
    precip_features_2019.append(col)
  elif 'precip 2014' in col:
    precip_features_2015.append(col)
  elif 'precip 2015' in col:
    precip_features_2015.append(col)
```

```python
# Separating the train dataset from the main dataframe
#
train = df[df.columns.difference(precip_features_2019)]

# Previewing the first two rows of the train dataset
#
train.head(2)
```

```python
# Separating the test dataset from the main dataframe
#
precip_features_2019.extend(['X',	'Y',	'elevation', 'LC_Type1_mode',	'Square_ID'])
test = df[precip_features_2019]

# Previewing the first two rows of the test dataset
#
test.head(2)
```

#### Renaming columns

```python
# Creating a dictionary of column names to be renamed for the training dataset
# The column names are renamed for conveniency
#
new_2015_cols = {}
for col, number in zip(precip_features_2015, range(1, len(precip_features_2015) + 1)):
  if 'precip' in col:
    new_2015_cols[col] = 'week_' + str(number) + '_precip'

    
# Creating a dictionary of column names to be renamed for the testing dataset
#
new_2019_cols = {}
for col, number in zip(precip_features_2019, range(1, len(precip_features_2019) + 1)):
  if 'precip' in col:
    new_2019_cols[col] = 'week_' + str(number) + '_precip'
    
# Renaming the columns
#
train.rename(columns = new_2015_cols, inplace = True)
test.rename(columns = new_2019_cols, inplace = True)
```

```python
# Previewing the first three rows of the cleaned train set
#
train.head(3)
```

#### Re-aligning the Train and Test Datasets

```python
# Separating the target variable
#
target = train.target_2015


# Aligning the training and testing datasets
#
train, test = train.align(test, join = 'inner', axis = 1)


# Previewing the first three rows of the cleaned and realigned test set
#
test.head(3)
```

```python
train.head(2)
```

### Feature Engineering

```python
from sklearn.preprocessing import KBinsDiscretizer
```

```python
train['XY']= train['X']*train['Y']
test['XY']= test['X']*test['Y']
```

```python
#discretisation of features to get 10 classsesofthe features

disc = KBinsDiscretizer(n_bins=10, encode='ordinal', strategy='uniform')

disc.fit(train[['elevation','week_7_precip','week_9_precip','week_11_precip','week_14_precip','week_15_precip','XY']])
```

```python
#disc.bin_edges_
```

```python
train_t = disc.transform(train[['elevation','week_7_precip','week_9_precip','week_11_precip','week_14_precip','week_15_precip','XY']])

train_t = pd.DataFrame(train_t, columns = ['elevations', 'week_7_precips','week_9_precips','week_11_precips','week_14_precips','week_15_precips','XYs'])

train_t.head(3)
```

```python
test_t = disc.transform(test[['elevation', 'week_7_precip','week_9_precip','week_11_precip','week_14_precip','week_15_precip','XY']])

test_t = pd.DataFrame(test_t, columns = ['elevations', 'week_7_precips','week_9_precips','week_11_precips','week_14_precips','week_15_precips','XYs'])
test_t.head(3)
```

```python
# merging of the discretization dataframe and original dataframe
trains= pd.merge(train, train_t, right_index=True, left_index=True)
test2= pd.merge(test, test_t, right_index=True, left_index=True)
test2.head(2)
```

```python
#drop some hindering columns/feature selection. 

to_drop=['Square_ID','week_1_precip','week_2_precip', 'week_3_precip','week_4_precip','week_5_precip','week_6_precip','week_8_precip','week_10_precip',
         'week_12_precip','week_13_precip','week_16_precip','week_17_precip']

X= trains.drop(to_drop, axis=1)
tests2= test2.drop(to_drop, axis=1)
y = target
```

```python
X.head(2)
```

```python
tests2.head(2)
```

```python
from sklearn.model_selection import train_test_split

x_train2, x_test2, y_train2, y_test2 = train_test_split(X, y, test_size=0.15, random_state=1001)
```

```python
model2 =  CatBoostRegressor(iterations=2000,
                             learning_rate=0.025,
                            
                             depth=6,
                             l2_leaf_reg=3.8,
                             metric_period= 500)

model2.fit(x_train2, y_train2,
                 eval_set=(x_test2, y_test2),
                 #cat_features=categorical_features_indices,                
                 use_best_model=True)
```

```python
#to drop for second model.it performed well without columns XY and xYs
to_drop2=['Square_ID','week_1_precip','week_2_precip', 'week_3_precip','week_4_precip','week_5_precip','week_6_precip','week_8_precip','week_10_precip','week_12_precip','week_13_precip','week_16_precip','week_17_precip',"XY",'XYs']

X2= trains.drop(to_drop2, axis=1)
tests= test2.drop(to_drop2, axis=1)
y = target
```

```python
X2.head(3)
```

```python
tests.head(2)
```

```python
x_train1, x_test1, y_train1, y_test1 = train_test_split(X2, y, test_size=0.15, random_state=42)
```

```python
model1 =  CatBoostRegressor(iterations=2000,
                             learning_rate=0.025,
                             depth=6,
                            l2_leaf_reg=3.8,
                             metric_period= 500)

model1.fit(x_train1, y_train1,
                 eval_set=(x_test1, y_test1),
                 #cat_features=categorical_features_indices,                
                 use_best_model=True)
```

### making submission

```python
sub= tests
sub['target_2019a'] = model1.predict(sub)
```

```python
sub2= tests2
sub2['target_2019b'] = model2.predict(sub2)
```

```python
sub['target_2019b']= sub2['target_2019b']
sub['target_2019']=(sub['target_2019a']+sub['target_2019b'])/2
sub.head(20)
```
